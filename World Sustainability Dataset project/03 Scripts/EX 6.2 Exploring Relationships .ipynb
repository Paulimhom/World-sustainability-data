### 6 Categorical Plot

```python
# Use a histogram to visualize the distribution of the variables.This way, we can determine categories for a certain ranges. 
# It would be wise to end up with too few observations in any one of the categories.
# The argument "kde" is added to a line that encompasses the distribution
sns.histplot(WorldSustainabilityDataset_NE['CO2 emissions in million tonnes'], bins = 20, kde = True)
```

```python
WorldSustainabilityDataset_NE.loc[WorldSustainabilityDataset_NE['CO2 emissions in million tonnes'] <= 25, 'Emission category'] = 'lowestest emmitters'
```

```python
WorldSustainabilityDataset_NE.loc[(WorldSustainabilityDataset_NE['CO2 emissions in million tonnes'] > 25) & (WorldSustainabilityDataset_NE['CO2 emissions in million tonnes'] <= 50), 'Emission category'] = 'lower average emmitters'
```

```python
WorldSustainabilityDataset_NE.loc[(WorldSustainabilityDataset_NE['CO2 emissions in million tonnes'] > 50) & (WorldSustainabilityDataset_NE['CO2 emissions in million tonnes'] <= 75), 'Emission category'] = 'Average emmitters'
```

```python
WorldSustainabilityDataset_NE.loc[WorldSustainabilityDataset_NE['CO2 emissions in million tonnes'] > 75, 'Emission category'] = 'Highest emmitters'
```
